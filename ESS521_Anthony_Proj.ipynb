# Using richdem for terrain attributes

```python
seak_gdal = gdal.Open(seak_fn)
seak_gdal_get = seak_gdal.GetRasterBand(1)
seak_np = seak_gdal_get.ReadAsArray()
```

```python
seak_np
```

```python
seak_rd  = rd.rdarray(seak_np, no_data=-9999)
seak_slope = rd.TerrainAttribute(seak_rd, attrib='slope_riserun')
rd.rdShow(seak_slope, axes=False, cmap='jet', figsize=(8,5.5))
```
